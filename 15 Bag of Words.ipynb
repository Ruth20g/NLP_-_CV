```python
import pandas as pd
import numpy as np
import re
from nltk.tokenize import word_tokenize
```

```python
sentence_1 = "It is a good practice for us."
sentence_2 = "It was also good to know about it."
```

```python
tokens= word_tokenize(sentence_1.lower()) + word_tokenize(sentence_2.lower())
```

```python
tokens
```

```python
tokens= set(tokens)
```

```python
tokens
```

```python
df= pd.DataFrame({}, index=[1,2], columns=list(tokens))
```

```python
df
```

```python
tokens1= word_tokenize(sentence_1.lower())
tokens2= word_tokenize(sentence_2.lower())
```

```python
tokens1
```

```python
tokens2
```

```python
counts1= [tokens1.count(x) for x in df.columns]
counts2= [tokens2.count(x) for x in df.columns]
```

```python
df.iloc[0,:]= counts1
df.iloc[1,:]= counts2
```

```python
df
```

# Use scikit learn

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cvt = CountVectorizer()
```

```python
new_data = cvt.fit_transform([sentence_1, sentence_2])
```

```python
new_data
```

```python
data = new_data.toarray()
```

```python
col = cvt.get_feature_names_out()
col
```

```python
df1 = pd.DataFrame(data,columns=list(col))
df1
```
